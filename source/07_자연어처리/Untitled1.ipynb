# 1.자연어처리
- 자연어: 일상적인 언어
- 자연어 처리 분야
    * 자연어 이해 : 형태소 분석 -> pos tagging(품사태깅) -> 의미분석(빈도분석, 연관분석)
    * 자연어 이해&생성 : RNN/LSTM/GRU -> seq2seq -> 트랜스포머
- 활용분야 : 트랜드분석, 탐색적 분석, 맞춤법검사, 번역기, 챗봇
# 2. 자연어 이해/생성 처리 절차
- 전처리: 단어/어절 추출 (특수문자 ㅎㅎ 제외)
- 분석후보 생성: 형태소 분리, 품사태깅
- 제약조건 : 불용어 처리, 규칙 확인
- 분석: 시각화(워드클라우드, text), 유사도 분석, 연관분석, RNN/LSTM/GUR, ...
# 3. 형태소 분석 엔진
