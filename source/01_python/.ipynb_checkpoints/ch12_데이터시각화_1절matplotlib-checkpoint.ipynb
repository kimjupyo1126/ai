## 1.2 그래프 그리기
- matplotlib으로 그래프그리기 위한 단계

    4) 그래프 커스터마이징 <br>
    5) 그래프 출력 및 저장 <br>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3] #2) 그래프 객체 생성사이즈는 인치
y = [-2, 4, 6] #3)
plt.figure(figsize=(7, 2)) #4) 커스터마이징
plt.plot(x, y)
plt.ylabel('some number')
plt.show() # 5)출력
plt.savefig('data/ch12.png', bbox_inches='tight') #파일출력
```

## 1.3그래프 영역 나누기
### 1) subplot() 함수로 서브플롯 추가

```python
import numpy as np
x = np.arange(0, 10, 0.1) # np.linspace(0, 9.9, 100)
x.shape
```

```python
plt.figure(figsize=(5,5))

plt.subplot(2, 1, 1) # 2행1열로 subplot에서 1번째 subplot은 다음과 같이 그린다
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 3) # 2행2열 subplot중 3번째 subplot
plt.plot(x, np.cos(x))
plt.subplot(224) # 2행2열 subplot 중 4번째 subplot
plt.scatter(x, np.sin(x)*np.cos(x))
plt.show()
```

```python
plt.subplot(2, 2, 1); plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2); plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3); plt.plot(x, np.sin(x))
plt.subplot(2, 2, 4); plt.scatter(x, np.sin(x)*np.cos(x))
plt.show()
```

## 2) subplots()함수로 서브플롯 추가
### plot함수에 색상이나 스타일 조정

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], 'r')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], 'g')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], 'b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
