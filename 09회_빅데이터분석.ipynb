정답: ① 1/2

설명: $E[X] = (0 \times 1/4) + (1 \times 1/2) + (2 \times 1/4) = 1$입니다. 분산 $V(X) = E[(X-1)^2]$는 $(0-1)^2 \times 1/4 + (1-1)^2 \times 1/2 + (2-1)^2 \times 1/4 = 1/4 + 0 + 1/4 = 1/2$이 됩니다.

설명: 제시된 그림처럼 오른쪽으로 꼬리가 긴(Right skew) 분포는 왜도가 양수($>0$)입니다. 이때 값의 크기는 왼쪽 봉우리에 위치한 최빈값이 가장 작고, 그다음 중앙값, 꼬리 쪽으로 끌려가는 평균값이 가장 큽니다.

# 33. 첨도와 이상값에 대한 설명으로 옳은 것은?

정답: ① 첨도는 데이터 평균값에 중요한 역할을 한다.

설명: 첨도(Kurtosis)는 분포의 뾰족한 정도와 꼬리의 두터움을 나타내며, 이는 데이터가 중앙(평균) 근처에 얼마나 집중되어 있는지를 보여줍니다. 4번 보기에서 이상값은 평균값에 큰 영향을 줍니다.

# 34. 다음 데이터의 선정 기준에 해당하는 것으로 알맞은 것은?

정답: ② 횡단면적 데이터(Cross-Sectional Data)

설명: 표를 보면 특정 시점(단일 시점)에 여러 업종(공단, 금융 등)과 여러 지역(서울, 경기 등)의 데이터를 수집한 형태입니다. 시간의 흐름을 추적하는 것이 아니라 특정 시점의 단면을 분석하므로 횡단면 데이터입니다.

# 35. 다음 중 확률표본추출법(Probability Sampling)은?

# 36. 분포 성질에 대한 설명으로 옳은 것은?

설명: 카이제곱분포는 독립적인 표준정규분포를 따르는 변수들의 제곱 합으로 정의되며, 이때 더해지는 변수의 개수 k를 자유도로 가집니다. 1번 포아송 분포는 평균과 분산이 같습니다.

# 37. 모분산이 $\sigma^2$이고 표본분산이 $s^2$인 경우 표본분산의 MSE(Mean Squared Error)값은?

설명: 표본분산 $s^2$의 분산 공식인 $Var(s^2) = \frac{2\sigma^4}{n-1}$을 묻는 문제입니다. 불편추정량의 경우 MSE는 분산과 같으므로 2번이 답이 됩니다.

# 41. 지도학습의 사례로 가장 적절한 것은?

설명: 지도학습은 정답(라벨)이 있는 데이터를 학습하는 것으로, 회귀분석과 분류분석(랜덤포레스트 등)이 대표적입니다. 3번 K-Means와 4번 주성분분석은 대표적인 비지도학습입니다. (2번 랜덤포레스트도 지도학습이나, 가장 기초적이고 명확한 통계 사례로 회귀분석이 흔히 제시됩니다.)

# 42. 회귀분석에 대한 설명으로 옳지 않은 것은?

정답: ② $\beta_1$, $\beta_0$가 0이면 $y=1$일 때, 확률은 0이다.

설명: 회귀계수 $\beta_1$(기울기)과 $\beta_0$(절편)가 0이면 종속변수 $y$는 독립변수와 상관없이 오차항에 의해서만 결정되거나 상수값이 됩니다. 확률이 0이라는 단정적인 설명은 옳지 않습니다.

# 43. 다음 내용에서 제시된 값을 보고, 회귀분석과 관련된 설명으로 가장 옳지 않은 것은?

회귀식 $Y = \beta_0 + \beta_1 X_i + \epsilon_i$잔차제곱합(SSE)의 자유도는 8, 독립변수 수 $k=1$회귀제곱합(SSR)과 총제곱합(SST)이 주어져 있음F-값이 주어져 있음

정답: ② 표본수 $n = 10$이다.

설명: 단순회귀분석($k=1$)에서 잔차제곱합(SSE)의 자유도는 $n - k - 1$입니다. 문제에서 자유도가 8이라고 했으므로, $n - 1 - 1 = 8$을 풀면 $n = 10$이 맞습니다. (제시된 이미지의 선택지 중 틀린 것을 고르는 문제인데, 실제 계산값과 일치하므로 다른 선택지의 수치 오류를 확인해야 합니다. 일반적으로 $R^2 = SSR / SST$ 공식 등이 바르게 적용되었는지 체크가 필요합니다.)

# 44. 회귀분석 유의성 검정방법에 대한 설명으로 옳은 것은?

설명: 통계적 가설 검정에서 p-value가 설정한 유의수준($\alpha$)보다 작다는 것은 관측된 결과가 우연히 일어날 확률이 매우 낮다는 것을 의미합니다. 따라서 "회귀계수는 0이다"라는 귀무가설을 기각하고, 해당 회귀 모델이 통계적으로 유의미하다고 판단합니다.

# 45. 다중공선성을 측정하는 지표로 옳은 것은?

# 59. 다차원척도법(MDS)과 t-SNE에 대한 설명 중 옳지 않은 것은?

설명: MDS는 데이터 간의 거리를 보존하며 저차원으로 투영하는 기법입니다. t-SNE는 고차원에서의 지역적 구조(Neighbor)를 저차원에서도 잘 보존하도록 설계되어 시각화에 매우 유리한 기법입니다.

# 60. 지수평활법 수식에 대한 설명으로 옳지 않은 것은?
$F_{t+1} = \alpha Z_t + (1-\alpha) F_t$

정답: ④ 단순 지수평활법은 추세나 계절성을 반영한다.

설명: 단순 지수평활법(Simple Exponential Smoothing)은 추세가 없는 데이터에 적합하며, 최근 데이터에 더 큰 가중치를 두어 미래를 예측합니다. 추세나 계절성을 반영하기 위해서는 홀트-윈터스(Holt-Winters) 모델과 같은 확장된 지수평활법이 필요합니다.

# 61. 다음 베이지안 문제에서 불량이 발생했을 때, A공정에서 발생할 확률로 옳은 것은?
A 공정 점유율 = 50%, 불량률 = 1%B 공정 점유율 = 30%, 불량률 = 2%C 공정 점유율 = 20%, 불량률 = 3%

정답: ① 5/17

설명: 전체 불량이 발생할 확률 $P(Fail)$은 각 공정의 (점유율 $\times$ 불량률)의 합입니다.$P(Fail) = (0.5 \times 0.01) + (0.3 \times 0.02) + (0.2 \times 0.03) = 0.005 + 0.006 + 0.006 = 0.017$입니다.이 중 A공정에서 불량이 발생했을 확률은 $P(A|Fail) = P(A \cap Fail) / P(Fail) = 0.005 / 0.017 = 5/17$이 됩니다.

정답: ② (가) RNN, (나) CNN, (다) 강화학습

설명: (가) 영화 평론은 문장(시계열 데이터)이므로 RNN이 적합합니다. (나) 안면 인식은 이미지 처리이므로 CNN이 핵심 기술입니다. (다) 로봇팔 가동은 보상(Reward)을 통해 최적의 행동을 학습하는 강화학습(Reinforcement Learning)이 주로 사용됩니다.

# 63. 다음 보기에서 시퀀스-투-시퀀스(Seq-to-Seq)에 대한 설명으로 옳은 것은?

(가) 길이가 자유롭다. (나) 길이에 상관없이 품질은 동일하다. (다) 마지막에 컨텍스트 벡터가 만들어진다.

정답: ② 가, 다

설명: Seq-to-Seq 모델은 입력과 출력 시퀀스의 길이가 달라도 처리가 가능하며(가), 인코더가 입력 문장을 압축하여 마지막에 '컨텍스트 벡터'를 생성해 디코더로 넘깁니다(다). 하지만 문장이 너무 길어지면 정보 손실로 인해 품질이 저하되는 한계가 있어 이를 극복하기 위해 어텐션(Attention) 기법이 등장했습니다.

# 64. 딥러닝 알고리즘 중 생성자와 판별자가 경쟁하며 학습하는 것은?

정답: ③ GAN

# 65. 다음 중 텍스트 마이닝에 대한 설명으로 옳지 않은 것은?

정답: ④ Stemming으로 올바른 문장에 대해 나무 모형을 표현

설명: Stemming(어간 추출)은 단어의 원형을 찾는 과정이지 나무 모형(Tree structure)을 표현하는 기법이 아닙니다. 나무 모형은 주로 의사결정나무나 구문 분석(Parsing) 결과물에서 나타납니다.

# 66. 네트워크에서 노드들이 전반적으로 연결된 정도는?

정답: ② 밀도(Density)

설명: 네트워크 분석에서 밀도는 네트워크 내 전체 가능한 연결 수 대비 실제 존재하는 연결 수의 비율을 의미하며, 노드들이 얼마나 빽빽하게 연결되어 있는지를 나타냅니다.

# 67. 다음 중 평균절대백분율오차(MAPE)에 대한 설명으로 옳은 것은?
정답: ③ $\frac{1}{n} \sum_{i=1}^{n} |\frac{실측값 - 예측값}{실측값}| \times 100\%$

설명: MAPE(Mean Absolute Percentage Error)는 실제값 대비 오차의 절대값 비율을 평균 낸 것입니다. 수치 데이터의 크기에 영향을 받지 않고 오차의 정도를 백분율로 파악할 수 있는 지표입니다.

# 68. 혼동행렬에 대한 설명으로 옳지 않은 것은?

정답: ① 정밀도(Precision)는 2개 클래스 이상에서만 확인 가능하다.

설명: 정밀도는 이진 분류(Positive/Negative)에서도 매우 중요하게 사용되는 지표입니다. 특정 클래스 이상에서만 확인 가능한 것이 아니라, 모델이 특정 클래스라고 예측한 것 중 실제 정답 비율을 측정할 때 언제든 사용됩니다.

# 69. 모든 ROC 커브에서 x축은 오분류율(FPR), y축은 (가)이다. (가)는 무엇인가?

정답: ② 민감도(Sensitivity)

설명: ROC 커브는 x축에 1-특이도(FPR)를, y축에 민감도(Recall, TPR)를 두고 모델의 성능을 시각화한 곡선입니다.

# 70. 랜덤 모델과 비교하여 해당 모델의 성과가 얼마나 향상되었는지 각 등급별로 파악하는 그래프는?
